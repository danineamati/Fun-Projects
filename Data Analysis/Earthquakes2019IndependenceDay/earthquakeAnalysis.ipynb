# Goal: Analyze Data from the July 4 - 6, 2019 Earthquake

```python
# Import statements

# Data handling
import pandas as pd
import numpy as np

# Graphing
import matplotlib.pyplot as plt
```

```python
# Read in the data as a dataframe
df = pd.read_csv("query_CA.csv")
df.head()
```

```python
df_large = df.query("mag >= 5")
df_large.head()
```

## First we want to graph a histogram of the magnitudes

#### Theoretically it should be exponential-like

```python
df.hist("mag", figsize=(10,8))

print("Average Magnitude: ", np.mean(df["mag"]))
print("Standard Deviation in Magnitude: ", np.std(df["mag"]))

print("Rate Param estimated from mean: ", 1 / np.mean(df["mag"]))
print("Rate Param estimated from standard deviation: ", 1 / np.std(df["mag"]))

plt.show()
```

```python
df_large.plot(x = "depth", y = "mag", kind = "scatter", figsize=(10,8), s = 60, grid = True)

print("Average Magnitude: ", np.mean(df_large["mag"]))
print("Standard Deviation in Magnitude: ", np.std(df_large["mag"]))

print("Rate Param estimated from mean: ", 1 / np.mean(df_large["mag"]))
print("Rate Param estimated from standard deviation: ", 1 / np.std(df_large["mag"]))

plt.show()
```

#### Suppose that the rate parameter is given by the standard deviation 
(since the mean is very skewed; the data only has magnitude 2.5+)

```python
lamRate = 1 / np.std(df["mag"])

multiplier = 1140



df.hist("mag", bins = [i / 2 for i in range(5, 15)], figsize=(10,8), zorder=0)


for i in np.linspace(40, 70, 5):
    new_mult = multiplier * i
    
    xVals = np.linspace(min(df["mag"]), max(df["mag"]))
    yVals = np.exp(- lamRate * xVals) # This gives the probabilities
    yVals = yVals * new_mult

    plt.plot(xVals, yVals, zorder=10, label = i)

    xSparse = np.array([i / 2 + .25 for i in range(5, 15)])
    ySparse = np.exp(- lamRate * xSparse) # This gives the probabilities
    ySparse = ySparse * new_mult

    plt.scatter(xSparse, ySparse, c = 'k', zorder=20)

plt.title("Number of Earthquakes Per 0.5 Magnitude Bin")
plt.xlabel("Magnitude")
plt.ylabel("Number Observed")
plt.legend()

plt.show()
```
